```python
import pandas as pd

copa_mundo = pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt', delimiter='\t')
copa_mundo
```

Usualmente, se concidera a Brasil como el mejor equipo del mundo debido a que es poseedor de 5 copas mundiales. Sin mebargo, el segundo lugar esta debatido, pues tanto Italia como Alemania tienen 4 copas cada uno. Entonces, Quien es el segundo y tercer mejor equipo del mundo? por ello, teniendo a disposicion una base de datos de goles de los paises a lo largo de la historia de mundiales, me gustaria conocer que equipo tienen mas goles a favor, si Italia o Alemania y de esta forma decidir quien tiene el segundo y tercer lugar como los mejores equipos del mundo.   

```python
#Conocer los goles de Italia a lo largo de la historia 
#Cuando es el equipo 1 
Italia_equipo1 = copa_mundo.loc[copa_mundo['equipo_1'] == 'Italia', 'equipo_1_final'].sum()
Italia_equipo1
```

```python
#Conocer los goles de Italia a lo largo de la historia 
#Cuando es el equipo 2
Italia_equipo2 = copa_mundo.loc[copa_mundo['equipo_2'] == 'Italia', 'equipo_2_final'].sum()
Italia_equipo2
```

```python
#Conocer los goles de Italia a lo largo de la historia 
Total_goles_italia= Italia_equipo1 + Italia_equipo2
print(f"Italia ha marcado un total de {Total_goles_italia} goles en la historia de los mundiales.")
```

```python
#Conocer los goles de Alemania a lo largo de la historia 
Alemania_equipo1 = copa_mundo.loc[copa_mundo['equipo_1'] == 'Alemania', 'equipo_1_final'].sum()
Alemania_equipo2 = copa_mundo.loc[copa_mundo['equipo_2'] == 'Alemania', 'equipo_2_final'].sum()
Total_goles_alemania= Alemania_equipo1 + Alemania_equipo2
print(f"Alemania ha marcado un total de {Total_goles_alemania} goles en la historia de los mundiales.")
```

```python
# Construccion de variables 
equipos = ['Italia', 'Alemania']
goles = [Total_goles_italia, Total_goles_alemania]
```

```python
#Construccion Grafico 1: Un solo grafico comparando los 2 paises 
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(equipos, goles, color=['blue', 'gray'])
plt.title('Quien tiene mas goles (1930-2018)')
plt.ylabel('Cantidad de Goles')
plt.xlabel('Equipo')
plt.tight_layout()
# Guardar el gráfico en formato PNG
Grafico1 = "comparacion_goles_italia_alemania.png"
plt.savefig(Grafico1, dpi=600) 
plt.show()
print(f"Gráfico guardado como {Grafico1}")
```

```python
#Gráfico 2: Dos graficos comparando otras variables 

import matplotlib.pyplot as plt

# vectores
tipo_equipo = ['Goles como equipo 1', 'Goles como Equipo 2']
goles_italia = [Italia_equipo1, Italia_equipo2]
goles_alemania = [Alemania_equipo1, Alemania_equipo2]

# Crear los subgráficos

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Configuración del espacio entre barras
ancho_barra = 0.35
pos = range(len(tipo_equipo))

# Primer subgráfico: Italia
axs[0].bar(pos, goles_italia, width=ancho_barra, color=['green', 'purple'], align='center')
axs[0].set_title('Italia')
axs[0].set_xticks(pos)
axs[0].set_xticklabels(tipo_equipo)

# Segundo subgráfico: Alemania
axs[1].bar(pos, goles_alemania, width=ancho_barra, color=['green', 'purple'], align='center')
axs[1].set_title('Alemania')
axs[1].set_xticks(pos)
axs[1].set_xticklabels(tipo_equipo)

# Título general, nombres de ejes y leyendas
fig.suptitle('Goles como equipo de ida y vuelta')
axs[0].set_ylabel('Cantidad de Goles')
for ax in axs:
    ax.set_xlabel('Tipo de equipo')

# Ajuste del grafico
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Guardar el gráfico en formato PNG 
grafico2 = "comparacion_golesIdaVenida_italia_alemania.png"
fig.savefig(grafico2, dpi=300) 

# Mostrar 
plt.show()

print(f"Gráfico guardado como {grafico2}")
```

Considerando que el mejor equipo es aquel que mas copas ha levantado y que lo ha logrado debido a la cantidad de goles que ha anotado, puedo decir que Italia es el segundo mejor equipo de la historia debido a que posee 4 copas del mundo y un total de 133 goles anotados por sobre Alemania que a pesar de que ha levandato 4 copas del mundo  ha anotado un total de 91 goles, 42 goles menos que Italia, lo que representa un 31% menos de anotaciones.
